```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df1=pd.read_csv('statewise.csv')
df1.head()
```

```python
df1.describe()
```

```python
df1.info()
```

```python
df2=pd.read_csv('odisha.csv')
df2.head()
df2.info()
```

```python
print('df1 shape : ',df1.shape)
print('df2 shape : ',df2.shape)
```

```python
subset=df1[63:]
print(subset)
```

```python
plt.figure(figsize=(6,6))
sns.barplot(data=subset,y='cityName', x='category',palette='viridis')
plt.ylabel('Clinic Type',fontsize=12,family='serif')
plt.xlabel('City', family='serif',fontsize=12,labelpad=20)
plt.xticks(family='serif')
plt.yticks(family='serif')
plt.title(label='Clinics in different cities',weight=250,family='serif')
plt.show()
```

```python
subset=df2[:30]
print(subset)
```

```python
plt.figure(figsize=(6,6))
sns.barplot(data=df2,y='District', x='Performance',palette='viridis')
plt.ylabel('District',fontsize=12,family='serif')
plt.xlabel('Performance', family='serif',fontsize=12,labelpad=20)
plt.xticks(family='serif')
plt.yticks(family='serif')
plt.title(label='Performance of clinics in Odisha in different districts',weight=250,family='serif')
plt.show()
```

```python
performance_to_size={
    'Minimum': 350,
    'Maximum': 800
}
performance_to_color={
    'Minimum': 'blue',
    'Maximum': 'red'
}
sizes=[performance_to_size[perf] for perf in df2['Performance']]
color=[performance_to_color[perf] for perf in df2['Performance']]
plt.figure(figsize=(10,6))
plt.scatter(x=df2['Total'],y=df2['District'],s=sizes,c=color,alpha=0.5)
plt.xlabel('Total Facilities Considered')
plt.ylabel('District')
plt.title('Bubble Chart')
plt.show()
```
